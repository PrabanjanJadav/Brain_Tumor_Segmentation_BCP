```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
file_path = "./CT_AVM.nii.gz"  
mri = nib.load(file_path).get_fdata()
print('MRI Shape:', mri.shape)
# slices = [mri[:, :, i] for i in range(5)]
for thing in range(70,80):
    plt.imshow(mri[:,  thing, : ], cmap='gray')
    plt.title(f'Slice {thing}')
    plt.axis('off')
    # plt.savefig(f'slice_{thing}.png')
    plt.show()
def plot_slices(mri, indices):
    for idx in indices:
        plt.imshow(mri[:, :, idx], cmap='gray')
        plt.title(f'Slice {idx}')
        plt.axis('off')
        plt.show()
def norm_all_slices(mri):
    l=mri.shape[2]
    for slice_index in range(l):
        norm_slice = [(mri[:, :, slice_index] - np.min(mri)) / (np.max(mri) - np.min(mri))]
    return norm_slice
```
